```python
#-----------------------------------Observational-----------------------------------#
#importing all the essential library # from this line!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split
#Patient Registries##
observational = pd.read_csv('Observational_cleaned_python.csv', sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
observational.head()
```

```python
del observational['nct_id']
observational.dtypes
```

```python
#Getting all the categorical columns except the overall status
categorical_columns3 = observational.select_dtypes(exclude = 'number').drop('overall_status', axis = 1).columns
observational.info()
```

```python
pd.crosstab(observational['overall_status'], observational['lead_Sponsor'])
```

```python
# Import the function
from scipy.stats import chi2_contingency
#Testing the relationship
chi_res = chi2_contingency(pd.crosstab(observational['overall_status'], observational['lead_Sponsor']))
print('Chi2 Statistic: {}, p-value: {}'.format(chi_res[0], chi_res[1]))
```

```python
# Import the function
from scipy.stats import chi2_contingency

chi2_check = []

for i in categorical_columns3:
    if chi2_contingency(pd.crosstab(observational['overall_status'], observational[i]))[1] < 0.05:
        chi2_check.append('Reject Null Hypothesis')
    else:
        chi2_check.append('Fail to Reject Null Hypothesis')
res = pd.DataFrame(data = [categorical_columns3, chi2_check] 
             ).T 
res.columns = ['Column', 'Hypothesis']
print(res)
```

```python
property_dummies = pd.get_dummies(data = observational[['overall_status', 'lead_Sponsor']], columns = ['overall_status'])
property_dummies
```

```python
pd.crosstab(property_dummies['lead_Sponsor'], property_dummies['overall_status_Completed'])
```

```python
check = {}
for i in res[res['Hypothesis'] == 'Reject Null Hypothesis']['Column']:
    dummies = pd.get_dummies(observational[i])
    bon_p_value = 0.05/observational[i].nunique()
    for series in dummies:
        if chi2_contingency(pd.crosstab(observational['overall_status'], dummies[series]))[1] < bon_p_value:
            check['{}-{}'.format(i, series)] = 'Reject Null Hypothesis'
        else:
            check['{}-{}'.format(i, series)] = 'Fail to Reject Null Hypothesis'
res_chi_ph = pd.DataFrame(data = [check.keys(), check.values()]).T
res_chi_ph.columns = ['Pair', 'Hypothesis']
res_chi_ph
#res_chi_ph.to_csv("Capstone.ChiTest_PR.csv")
```
