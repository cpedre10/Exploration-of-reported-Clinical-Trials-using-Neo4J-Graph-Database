```python
#-----------------------------------Patient Registries-----------------------------------#
#importing all the essential library # from this line!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split
#Patient Registries##
registries = pd.read_csv('PatientRegistry_cleaned_final.csv', sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
registries.head()
```

```python
del registries['nct_id']
registries.dtypes
```

```python
registries.describe(include='all')
```

```python
#Getting all the categorical columns except the overall status
categorical_columns2 = registries.select_dtypes(exclude = 'number').drop('overall_status', axis = 1).columns
registries.info()
```

```python
pd.crosstab(registries['overall_status'], registries['lead_Sponsor'])
```

```python
# Import the function
from scipy.stats import chi2_contingency
#Testing the relationship
chi_res = chi2_contingency(pd.crosstab(registries['overall_status'], registries['lead_Sponsor']))
print('Chi2 Statistic: {}, p-value: {}'.format(chi_res[0], chi_res[1]))
```

```python
# Import the function
from scipy.stats import chi2_contingency
chi2_check = []
for i in categorical_columns2:
    if chi2_contingency(pd.crosstab(registries['overall_status'], registries[i]))[1] < 0.05:
        chi2_check.append('Reject Null Hypothesis')
    else:
        chi2_check.append('Fail to Reject Null Hypothesis')
res = pd.DataFrame(data = [categorical_columns2, chi2_check] 
             ).T 
res.columns = ['Column', 'Hypothesis']
print(res)
```

```python
property_dummies = pd.get_dummies(data = registries[['overall_status', 'primary_purpose']], columns = ['overall_status'])
property_dummies
```

```python
pd.crosstab(property_dummies['primary_purpose'], property_dummies['overall_status_Completed'])
```

```python
check = {}
for i in res[res['Hypothesis'] == 'Reject Null Hypothesis']['Column']:
    dummies = pd.get_dummies(registries[i])
    bon_p_value = 0.05/registries[i].nunique()
    for series in dummies:
        if chi2_contingency(pd.crosstab(registries['overall_status'], dummies[series]))[1] < bon_p_value:
            check['{}-{}'.format(i, series)] = 'Reject Null Hypothesis'
        else:
            check['{}-{}'.format(i, series)] = 'Fail to Reject Null Hypothesis'
res_chi_ph = pd.DataFrame(data = [check.keys(), check.values()]).T
res_chi_ph.columns = ['Pair', 'Hypothesis']
res_chi_ph
#res_chi_ph.to_csv("Capstone.ChiTest_PR.csv")
```
